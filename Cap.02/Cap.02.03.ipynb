## Step 3:- Create Visualizations

```python
import matplotlib.pyplot as plt
```

```python
# Check the count of galaxy, star and qasar in our data

# Adjusting the Figure Size
plt.figure(figsize = (6, 5))

''' We can plot directly using pandas '''
# value_counts() gives counts of unique values in the column it is used on
ax = df['class'].value_counts().plot(kind = 'bar', color = 'green') # Activity:- how to color the bars here?

# Using ax (axis) to display the bar_label using containers[0] as thats where the location is of the counts
ax.bar_label(ax.containers[0], color='maroon', weight='bold')

# Customizing the Plot a bit more
plt.xticks(weight='bold', rotation=0)
plt.title('Visualizing the distribution of Classes in the Data', color='r', weight='bold', fontsize=12)

# Save the Figure and Display the output
plt.savefig('ClassDistribution.png') # You can download the visualization from files section (left panel)
plt.show()
```

### **Scikit-Image**

* Add any image to Google Colab (I will use the one I have for this session)
* How to read the image using skimage
* How to visualize the image array
* How to use slicing to crop the image

```python
# Get the imread from skimage's io
from skimage.io import imread
```

```python
# Get the image path from session storage of colab
img_path = "https://github.com/MauricioRPF/Astronomy-Machine-Learning/blob/main/Assets/Moon-Pleiades-AirplaneLR.jpg?raw=true"
```

```python
# Read the image path and convert it to array of pixels
img_array = imread(img_path)
```

```python
# What is the shape of this image?
print(img_array.shape)
```

**Understand the shape of this image_array**:-

* Axis 0:- Rows
* Axis 1:- Cols
* Axis 3:- Channels

* If channels = 3, then it represents RGB (Red, Green, and Blue Channels)

* If channel = 1 (or only 2D array) then it represents a gray scale output.


```python
# Check the image array
print(img_array)
```

```python
# What are minimum and maximum values in this array?
print(f'Min:- {img_array.min()}\nMax:- {img_array.max()}')
```

```python
# This is uint-8 datatype - crosscheck it
img_array.dtype
```

```python
# Visualize the image
plt.imshow(img_array, cmap = 'viridis')
plt.show()
```

```python
# Grayscale output
plt.imshow(img_array[:, :, 2], cmap='gray') # 0 - red, 1 - green, 2 - blue
plt.show()
```

```python
# Slice the moon and the pleiades constellation
constellation = img_array[60:120, 50:140, :]
moon = img_array[185:233, 214:264, :]

# Tuple to store the data for plotting
plot_data = (
    (constellation, moon),
    ('Pleiades', 'Moon'),
    ('blue', 'gray')
)


# Using for loop to loop through the tuple to create subplot
for index, (data, title, color) in enumerate(zip(*(plot_data))):
  plt.subplot(1, 2, index+1)
  plt.imshow(data)
  plt.title(title, color = color)
  plt.axis('off')


# Adjust the layout and display the output
plt.tight_layout()
plt.show()
```
